### lineplot

```python
# 연속형 샘플 데이터 생성 -> 시각화에 사용
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([1, 2, 3, 4, 5, 6])

# matplotlib lineplot
plt.plot(x, y, '--')
plt.plot(x, y*2)
plt.show()
```

### histogram
- 히스토그램은 연속형 변수의 값/분포를 보여주는 도구

## 시각화 실습
- seaborn 라이브러리 tips dataset

```python
tips = sns.load_dataset('tips')
tips
```

### scatterplot
-  총 청구 금액과(total_bill) 팁(tip) 사이의 관계를 나타내는 산점도를 그리고 관계 묘사

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)

# 우상향 그래프(양의 상관관계)
```

```python
# 두 변수(total_bill, tips)의 상관계수

import numpy as np

print(np.corrcoef(tips['total_bill'], tips['tip']))
print()
print(np.corrcoef(tips['total_bill'], tips['tip'])[0, 1])

# 0.68에 가까운 상관계수를 가지는 강한 양의 상관관계
```

```python
# regression plot
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
##### time 변수를 기준으로 점의 색깔을 구분 #####

# seaborn scatterplot(hue='time')
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

```python
# seaborn regplot(hue 옵션 x) -> 데이터 전처리 필요

# time 변수의 unique 값들을 분할하여 저장
tip_1 = tips.loc[tips['time'] == 'Lunch']
tip_2 = tips.loc[tips['time'] == 'Dinner']

sns.regplot(x='total_bill', y='tip', data=tip_1)
sns.regplot(x='total_bill', y='tip', data=tip_2)
plt.show()
```

```python
# seaborn lmplot
sns.lmplot(x='total_bill', y='tip', hue='time', height=4, data=tips)
```

```python
# day 변수를 기준으로 점의 색깔을 구분 -> scatterplot
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips)
```

### boxplot
- 박스-휘스커 플롯(Box-Whisker Plot) 또는 간단히 박스 플롯
- 박스와 박스 바깥의 선(whisker)으로 구성

- boxplot: http://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
# matplotlib boxplot(연속형 변수)
plt.boxplot(tips['tip'])
plt.show()
```

```python
# 이상치(outlier) 모양과 색깔, xlabel, title 설정
plt.boxplot(tips['tip'], sym='bo') # bo -> b는 color, o는 marker
plt.title('box plot of tip')
plt.show()
```

```python
# 두 변수(total_bill, tip)에 대한 boxplot 동시 출력
plt.boxplot([tips['total_bill'], tips['tip']], sym='bo')
plt.title('box plot of tip')
plt.show()
```

```python
# seaborn boxplot

# 여러 그룹에 대한 boxplot -> 하나의 연속형 변수를 다수의 그룹으로 분할
# 팁(tip)에 대한 요일(day)별 boxplot
sns.boxplot(x='day', y='tip', data=tips)
plt.show()
```

```python
# 성별(sex) 그룹 추가
# 요일(day)별 팁(tips)을 다시 성별(sex)로 구분
sns.boxplot(data=tips, x='day', y='tip', hue='sex')
plt.show()
```

```python
# 그래프 크기 조정
plt.figure(figsize=(8, 5))

sns.boxplot(x='day', y='tip', hue='sex', data=tips)
plt.show( )
```

```python
# 남녀 팁(tip) 구분하여 비교
sns.boxplot(x='sex', y='tip', data=tips)
plt.show()
```

### violinplot
- 일변량(종속변수 1개), 연속형 데이터를 다루는 바이올린 모양의 차트
- column에 대한 데이터의 비교 분포도를 확인
- 둥근 부분(뚱뚱한 부분)은 데이터의 분포를 표현
- 양쪽 끝 뾰족한 부분은 데이터의 최솟값/최댓값을 표현
 - 상자 그림(box plot)이 중앙값, 표준편차 등 분포의 간략한 특성만을 보여준다면 violin plot, strip plot, swarm plot 등은 카테고리 값에 따른 데이터의 실제 분포와 전체 형상을 제공

```python
sns.violinplot(x='sex', y='tip', data=tips)
plt.show()
```

```python
# 요일(day)별 총 청구 금액(total_bill)을 흡연 여부(smoker)를 기준으로 구분
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='YlGn_r' )
plt.show()
```

```python
# 흡연 여부(smoker)를 기준으로 구분하되, 요일(day)별 하나의 그래프로 표현(split=True)
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted', split=True)
plt.show()
```

### 피벗 테이블(pivot table) 시각화

```python
tips
```

```python
# 피벗 테이블 생성(행 인덱스: day, 열 인덱스: size, values: tip)
# default aggfunc -> mean
pivot = tips.pivot_table(index='day', columns='size', values='tip')
pivot
```

### sns.jointplot
- 두 개의 수치형 변수 간의 관계

```python
titanic_df = pd.read_csv('titanic_train.csv')
```

```python
titanic_df.columns
```

```python
# 성별(Sex)과 생존 여부(Survived)의 관계를 시각화(countplot)
sns.countplot(x='Survived', hue='Sex', data=titanic_df)
```

```python
# 나이(Age)에 따라 탑승객을 분류하는 함수 생성

def get_category(age):
    
    if age <= -1: cat = 'Unknown'
    elif age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat
```

```python
# Age 컬럼에 get_category() 함수를 적용한 Age_cat 컬럼 생성
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
```

```python
plt.figure(figsize=(10, 6))
group_names = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Elderly']

# group_names 순서대로 barplot 출력
sns.barplot(x='Age_cat', y = 'Survived', hue='Sex',
            data=titanic_df, order=group_names)
```

```python
# seaborn violinplot
# 클래스(Pclass)별 연령(Age)의 분포를 성별(Sex)로 나누어 시각화
plt.figure(figsize=(10, 7))
sns.violinplot(data=titanic_df, x='Pclass', y='Age', hue='Sex', split=True)
```

```python
# 컬럼별 상관관계를 히트맵(heatmap)으로 시각화
plt.figure(figsize=(10, 7))
sns.heatmap(titanic_df.corr(), annot=True, cmap='Blues')
```
