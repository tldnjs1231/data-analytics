## 데이터프레임 합치기

### concat 연습문제

```python
import pandas as pd
```

```python
df4 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                   'b': ['b0', 'b1', 'b2', 'b3'],
                   'c': ['c0', 'c1', 'c2', 'c3']},
                   index=[0, 1, 2, 3])

df5 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                   'b': ['b2', 'b3', 'b4', 'b5'],
                   'c': ['c2', 'c3', 'c4', 'c5'],
                   'd': ['d2', 'd3', 'd4', 'd5']},
                  index=[2, 3, 4, 5])
```

```python
df4
```

```python
df5
```

```python
# row기준으로 합치고 result1에 저장
result1 = pd.concat([df4, df5])
result1
```

```python
# 인덱스를 다시 만들기 위해 기존 인덱스 reset
result1.reset_index(inplace=True)
```

```python
result1
```

```python
# index 컬럼 삭제 (result1 데이터 유지)
result1.drop('index', axis=1)
```

```python
# ignore_index = True를 사용하여 인덱스를 새로 생성한 후,
# df4, df5를 합친 결과를 result2에 저장
result2 = pd.concat([df4, df5], ignore_index=True)
result2
```

```python
# df4, df5를 컬럼 기준으로 합치고 result3에 저장
result3 = pd.concat([df4, df5], axis=1)
result3 # 컬럼명 중복
```

```python
# df4의 컬럼 rename, {'a':'a2', 'b':'b2', 'c':'c2'}하고 df4에 저장
df4 = df4.rename(columns={'a':'a2', 'b':'b2', 'c':'c2'})
df4
```

```python
# df4, df5를 컬럼 기준으로 concat하고 result4에 저장
result4 = pd.concat([df4, df5], axis=1)
result4
```

### Row 기준 합치기

```python
# 원본을 손실하지 않기 위해 df1의 복사본 df1_copy 생성
df1_copy = df1.copy()
```

```python
df1_copy
```

row 기준으로 합칠 때는 sort=False 옵션을 주어 순서가 유지되도록 합니다.

```python
# df1, df1_copy concat, sort옵션은 False(인덱스가 뒤죽박죽 되지 않도록)
pd.concat([df1, df1_copy], sort=False)
```

인덱스가 꼬이지 않게 하는 두가지 방법
- ignore_index =  True 
- reset_index(drop=True) : 
  인덱스초기화 할때 기존 인덱스가 컬럼화 되므로  drop=True 옵션으로 index column 이 생성되지 않도록 한다

```python
# 1)ignore_index=True을 사용하여 df1, df1_copy concat
pd.concat([df1, df1_copy], ignore_index=True)
```

```python
# 2) concat 후에 reset_index 하는 방법
df_concat = pd.concat([df1,df1_copy],sort=False)
df_concat
```

```python
# 인덱스를 리셋하고 drop = True 를 사용하여 원래 인덱스 삭제
df_concat.reset_index(drop=True)
```

### column 기준으로 합치기
- 컬럼기준으로 합칠때는  axis=1 옵션 사용

```python
# df1, df2 를 컬럼 기준으로 합치기
pd.concat([df1, df2], axis=1)
# 두 데이터의 행의 개수가 같아 null 값 없이 병합됨
```

```python
# 행의 개수가 맞지 않을 경우의 column concat을 확인
# df2에서 4, 11행을 삭제하고 df3에 저장
df3 = df2.drop([4, 11])
```

```python
df3
```

```python
# df1, df3 컬럼 기준으로 합친후 데이터가 어떻게 들어가 있는지 확인
pd.concat([df1, df3], axis=1)

# concat의 경우 값이 존재하지 않아도 양쪽의 데이터를 모두 출력 (존재하지 않는 값은 NaN 처리)
```

### Merge


```python
df1.head(3)
```

```python
df2.head(3)
```

df1과 df2는 '이름' 컬럼이 겹치므로 이름을 기준으로 병합 가능

#### left, right 방식

```python
# df2에서 1, 5, 10행을 삭제하고 df2_right에 저장
df2_right = df2.drop([1, 5, 10])
```

```python
df2_right
```

- 열 방향 concat의 경우 인덱스를 기준으로 병합하고 존재하지 않는 값은 NaN 처리하기 때문에,
- 기존의 인덱스가 남아있다면 행을 임의로 삭제하더라도 데이터가 섞이지 않아 merge, concat 방식에 차이가 없음
- 따라서 기존의 인덱스를 삭제하고 인덱스를 재배열하기 위해 reset_indxe(drop=True) 적용

```python
# df2_right에서 인덱스 재배열해서 df2_right에 저장
df2_right = df2_right.reset_index(drop=True) # drop=True: 'index' 열 생성 방지
```

```python
df2_right
```

```python
# 열 방향 concat
pd.concat([df1, df2_right], axis=1)

# df1, df2_right가 공통 열인 '이름' 기준이 아니라 인덱스 기준으로 병합되어 데이터가 섞임
```

```python
# 이름을 기준으로 merge하는데 방법은 how='left'
pd.merge(df1, df2_right, how='left', on='이름')

# 1, 5, 10행의 성별과 미디어지수는 NaN 처리되고 나머지 데이터가 잘 병합된 것 확인 가능
```

```python
# 이름을 기준으로 merge, how='right'
pd.merge(df1, df2_right, how='right', on='이름')

# df2_right에 존재하는 행만 이름을 기준으로 병합
```

- 컬럼명은 다르지만 동일한 컬럼일 경우의 merge
  - left_on, right_on 옵션사용

```python
df2_right.head(3)
```

```python
# df2_right 컬럼명(이름을 name으로)을 수정
df2_right.columns = ['name','성별','미디어지수']
```

```python
df2_right.head(3)
```

```python
# left_on='이름', right_on='name', how='outer'
df = pd.merge(df1, df2_right, left_on='이름', right_on='name', how='outer')
df
```
